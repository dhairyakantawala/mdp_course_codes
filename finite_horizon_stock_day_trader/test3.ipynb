```python
def t_prob(next_state, prev_state, action, t_matrix):
    if action == -2:
        prev_r, _ = prev_state
        next_r, _ = next_state
        return float(t_matrix.loc[prev_r, next_r])
    
    elif action == -1:
        prev_r, prev_w = prev_state
        next_r, next_w = next_state
        if (next_w - prev_w)!=-1:
            return 0.0
        else:
            return float(t_matrix.loc[prev_r, next_r])
        
    elif action == 0:
        prev_r, prev_w = prev_state
        next_r, next_w = next_state
        if (next_w - prev_w)!=0:
            return 0.0
        else:
            return float(t_matrix.loc[prev_r, next_r])

    elif action == 1:
        prev_r, prev_w = prev_state
        next_r, next_w = next_state
        if (next_w - prev_w)!=1:
            return 0.0
        else:
            return float(t_matrix.loc[prev_r, next_r])

    elif action == 2:
        prev_r, _ = prev_state
        next_r, _ = next_state
        return float(t_matrix.loc[prev_r, next_r])
```
